# Simple Queries

```python
#Find the subset of df where the client's last name is "Smith"
df=pd.read_csv('client_list_practice.csv')
df[df.LAST_NAME=='Smith']
```

```python
#Find the subset of df where the client's hair color is not black.
df[df.HAIR_COLOR!='black']
```

```python
#Find the subset of df where the client's hair color is red and reset the values to "ginger"
x= df[df.HAIR_COLOR == 'red'].index
df.loc[x,"HAIR_COLOR"] = 'ginger'
df 
```

# Complex Queries

```python
#Find the subset of df where the clients are females older than 30 years.
df[(df.GENDER=='F')&(df.AGE>30)]
```

```python
#Repeat problem 1, but return only the hair color and eye color.
#df[df.AGE > 30][["HAIR_COLOR","EYE_COLOR"]]
df[["HAIR_COLOR","EYE_COLOR"]][df.AGE > 30]
```

```python
#Find the unique combination of hair and eye color for women older than 25 years
df[(df.AGE > 25) & (df.GENDER == 'F')].HAIR_COLOR.unique()
```

```python
#Perform a merge using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as clients.

df=pd.read_csv('client_list.csv')
df4 = pd.read_csv('customer_id_list.csv')
clients = pd.merge(df,df4) 
```

```python
print(clients)
```

```python
#Perform a merge using clients and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as detailed_sales
df5 = pd.read_csv('purchase_log.csv')
detailed_sales = pd.merge(clients,df5)
detailed_sales
```

```python
#Use groupby to find the client who spent the most money on purchases. 
#Determine how much he/she spent. HINT: save the intermediate dataframe from using groupby as spenders before applying slicing to determine the client who spent the most money on purchases.
spenders = detailed_sales.groupby('PRICE').max()
spenders
```
